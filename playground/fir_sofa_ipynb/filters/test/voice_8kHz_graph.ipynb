```python
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import soundfile as sf

from scipy import signal


%matplotlib inline
```

```python
data, sr = sf.read("/tmp/white_noise_20.0s.wav")
```

```python
filter1 = signal.firwin(numtaps=512, cutoff=8000, width=None, 
                            window='hamming', pass_zero=True, 
                            nyq=None, fs=sr)
```

```python
y1 = signal.lfilter(filter1, 1, data)
```

```python
plt.plot(y1)
```

```python
sf.write(file="/tmp/white_noise_20.0s_scipy_8kHz_lpf.wav", data=y1, samplerate=48000, endian="LITTLE", format="WAV", subtype="PCM_16")
```
