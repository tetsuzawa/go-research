```python
satellites_info.Uplink = satellites_info.Uplink.replace(".","-")
satellites_info.Uplink = satellites_info.Uplink.replace(pd.np.nan,"-")

satellites_info.Downlink = satellites_info.Downlink.replace(".","-")
satellites_info.Downlink = satellites_info.Downlink.replace(pd.np.nan,"-")

satellites_info.Beacon = satellites_info.Beacon.replace(".","-")
satellites_info.Beacon = satellites_info.Beacon.replace(pd.np.nan,"-")
```

```python
i =0
for i in range(len(satellites_info["Mode"])):
    satellites_info.at[i,"Mode"] = satellites_info.at[i,"Mode"].replace('/',',').replace(" ,", ",").replace(", ",",").replace(" CW ", ",CW,").replace(" CW",",CW").replace("\n",",")
    
for i in range(len(satellites_info["tle"])):
    satellites_info.at[i,"tle"] = str(satellites_info.at[i,"tle"]).replace("nan", ".")
    
for i in range(len(satellites_info["tle_name"])):
    satellites_info.at[i,"tle_name"] = str(satellites_info.at[i,"tle_name"]).replace("nan", ".")
```

```python
def digits_judge(frequency):
    if len(frequency) >= 7:
        return frequency + (10-len(frequency))*" "
    elif len(frequency) == 6:
        return " " + frequency
    else:
        return frequency
```

```python
def equal_bar(i, col, Mode):
    if df.at[i, col] == "-":
        return "-" + "\n"
    else:
        return df.at[i, col] + Mode + "\n"
```
