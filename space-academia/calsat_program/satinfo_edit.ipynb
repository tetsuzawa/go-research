```python
def digits_judge(frequency):
    if len(frequency) >= 7:
        return frequency + (10-len(frequency))*" "
    elif len(frequency) == 6:
        return " " + frequency
    else:
        return frequency
```

```python
df["remote_mode"] = ""
```

```python
i = 0
for mods in df["Mode"]:
    mode_list =[]
    
    for mod in mods.split(","):
        
        if mod=="CW":
            mode_list.append("CW")
            
        elif mod=="nan":
            pass
            
        elif mod=="":
            pass
        else:
            mode_list.append("FM")
    df.at[i, "remote_mode"] = list(set(mode_list))
    i += 1
```

```python
def equal_bar(i, col, Mode):
    if df.at[i, col] == "-":
        return "-" + "\n"
    else:
        return df.at[i, col] + Mode + "\n"
```

```python
text = ""
for i in range(len(df)):
    if len(df.at[i,"remote_mode"]) > 1:
        for mode in df.at[i, "remote_mode"]:
            text += df.at[i,"elem_name"] + "[" + mode  + "]" + "\n" + equal_bar(i,"Uplink","FM") + equal_bar(i,"Downlink","FM") + equal_bar(i, "Beacon", mode) + df.at[i,"NORAD"] + "Mode" + df.at[i,"Mode"] + "\n" + df.at[i,"Callsign"] + "\n"
            #for j in range(len(df.columns)-1):
            #    text += df.iat[i, j]
            #text += mode
    
    else:
        for mode in df.at[i, "remote_mode"]:
            text += df.at[i,"elem_name"] + "\n" + equal_bar(i,"Uplink","FM") + equal_bar(i,"Downlink","FM") + equal_bar(i, "Beacon", mode) + df.at[i,"NORAD"] + "Mode:" + df.at[i,"Mode"] + "\n" + df.at[i,"Callsign"] + "\n"
        
text += "＠終わり(SATINFO)"
print(text)
```
