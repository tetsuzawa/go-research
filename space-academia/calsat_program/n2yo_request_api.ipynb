```python
import pandas as pd
import numpy as np
import time 
import codecs
import bs4
import requests
import json
```

```python
url_base = "https://www.n2yo.com/rest/v1/satellite/tle/"
api_key = "GGTNRF-RAJLVZ-FX7G6Q-45UG"
```

```python
def get_TLE(NORAD_ID):
    url = url_base + NORAD_ID + "&apiKey=" + api_key
    
    headers = {"content-type": "application/json"}
    
    try:
        res = requests.get(url, headers=headers)
        res.raise_for_status()
        data = res.json()
        return data
        
    except:
        return None
```

```python
data = get_TLE("44103")
print(json.dumps(data,indent=4))
print(data["info"]["satid"])
```
